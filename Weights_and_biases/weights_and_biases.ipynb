```python
import numpy as np
import matplotlib.pyplot as plt
```

## Understanding effect of weight and bias

##### Equation:
    Output = x * weights + bias

```python
# Function to visualize axes
def getPlot(x, weight, bias, figsize = (8,7)):
    fig = plt.figure(figsize= figsize)
    ax = fig.add_subplot(111)
    
    # Plotting function
    y = x * weight + bias
    ax.plot(x, y)
    ax.set_title(f"Weight: {weight}, Bias: {bias}")

    # Creating graph
    ax.spines['left'].set_position('zero')
    ax.spines['right'].set_color('none')
    ax.spines['bottom'].set_position('zero')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')
    ax.axhline(linewidth=1, color='grey')
    ax.axvline(linewidth=1, color='grey')
    ax.set_xlim(-400, 400)
    ax.set_ylim(-400, 400)
    ax.plot(0,bias, marker = "o", markersize = 10)
    ax.annotate(f"(0, {bias})",xy=(10,bias + 10))
```

```python
x = np.arange(-100,100)
weight = 10
bias = 0
getPlot(x, weight,bias)
```

```python
getPlot(x, 4, 0)
```

With increase in weights the slope of graph is increasing as weight = slope

```python
# Negative weight
getPlot(x,-4,0)
```

### Effect of Bias

```python
# Increasing bias (Positive bias)
getPlot(x,10, 40)
```

The y-intercept of graph or "the point at which output becomes 0" is shifted

```python
getPlot(x,7, -64)
```
